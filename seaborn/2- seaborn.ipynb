# Q1.Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.

```python
fmri=sns.load_dataset('fmri')
fmri
```

```python
fmri['region'].unique()
```

```python
fmri['event'].unique()
```

```python
plt.figure(figsize=(10,6))

sns.lineplot(x='timepoint', y='signal', data=fmri , hue='event', style='region')

plt.title("FMRI Signal over Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal")

plt.legend(title="Event & Region")
```

# Q2.Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.

```python
t=sns.load_dataset('titanic')
t
```

```python
fig , axes= plt.subplots(1,2 , figsize=(14,6))

# age vs pclass
sns.boxplot(data=t , x='pclass', y='age', ax=axes[0])
axes[0].set_title("Age Distribution by Passenger Class")
axes[0].set_xlabel("Passenger Class")
axes[0].set_ylabel("Age")

# fare vs pclass
sns.boxplot(data=t, x='pclass' , y='fare', ax=axes[1])
axes[1].set_title("Fare Distribution by Passenger Class")
axes[1].set_xlabel("Passenger Class")
axes[1].set_ylabel("Fare")
```

# Q3. Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
d=sns.load_dataset('diamonds')
d
```

```python
plt.figure(figsize=(14,6))
sns.displot(x='price', hue='cut', data=d , kde=True)

_=plt.title("Histogram of Diamond Prices by Cut")
plt.show()
```
