# Q1.Using the data plot a bar plot and a horizontal bar plot.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

#vertical
axes[0].bar(company, profit, color=['red', 'blue', 'green', 'purple'])
axes[0].set_title("Company Profits (Vertical)")
axes[0].set_xlabel("Company")
axes[0].set_ylabel("Profit ($)")
axes[0].bar_label(axes[0].containers[0], fmt='%.0f', padding=3)


# horizontal
axes[1].barh(company, profit, color=['red', 'blue', 'green', 'purple'])
axes[1].set_title("Company Profits (Horizontal)")
axes[1].set_xlabel("Profit ($)")
axes[1].set_ylabel("Company")
axes[1].bar_label(axes[1].containers[0], fmt='%.0f', padding=3)


plt.tight_layout()
plt.show()
```
