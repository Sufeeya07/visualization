```python
MATPLOTLIB -- PROVIDE SUMMARY OF DATA IN GRAPH
Matplotlib supports a wide range of plot types, including line plots, scatter plots, bar charts, histograms, pie charts, 3D plots, and more.
Matplotlib allows users to customize almost every aspect of a plot, including colors, line styles, markers, fonts, axes, labels, and annotations.
```

```python
graph  is representation of data in x y or z axis
```

```python
import matplotlib.pyplot as plt

# NumPy's powerful numerical computing capabilities to prepare, manipulate, and analyze data before visualizing it using Matplotlib. 
import numpy as np
```

```python
x= np.linspace(0,10, 200)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x,y)
```

```python
plt.xlabel("this is my 'x' data")
plt.ylabel("this is my 'y'data")
plt.title('this is whole graph ')
plt.plot(x,y )

# plt.show()
```

```python
# ---- use diferent types of plot

x1 = np.random.rand(50)
y1 = np.random.rand(50)
```

```python
x1
```

```python
y1
```

```python
col = np.random.rand(50)
sizes = 200 *np.random.rand(50)
plt.scatter(x1,y1 , c=col , s = sizes , alpha = .5)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y , '--y' , linewidth = '10' )
plt.xlabel("this is my 'x' data")
plt.ylabel("this is my 'y'data")
plt.title('this is whole graph ')
plt.grid()
```

```python
col = np.random.rand(50)
sizes = 200 *np.random.rand(50)
color = ['red'  ,'blue' ,'yellow']
plt.scatter(x1,y1 , c=col , s = sizes , alpha = .5 , marker = 'P')
```

```python
#-----3d cordinate

x5 = np.random.rand(50)
y5 = np.random.rand(50)
z5 = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x5,y5,z5)
plt.show()
```
